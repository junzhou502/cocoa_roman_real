```python
import getdist
import numpy as np
import time
t0 = time.time()

chains = [
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/chains/roman_lcdm_baseline_mcmc/',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/chains/roman_wcdm_baseline_mcmc/',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/lsst_real/chains/lsst_lcdm_baseline_mcmc/',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/lsst_real/chains/lsst_wcdm_baseline_mcmc/',
]

output = [
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/convergence_check_lcdm_baseline.txt',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/convergence_check_wcdm_baseline.txt',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/lsst_real/convergence_check_lcdm_baseline.txt',
    '/project/chihway/junzhou/cocoa_approx/Cocoa/projects/lsst_real/convergence_check_wcdm_baseline.txt'
]

for output_idx, chain in enumerate(chains):
    print(f'[{time.time()-t0:.2f}] {output_idx+1}/{len(chains)}: perparing....')
    samples = getdist.loadMCSamples(chain, settings={'ignore_rows':0.3, 'contours': [0.68]})
    param_names = [p.name for p in samples.getParamNames().names]
    with open(output[output_idx], 'w', encoding='utf-8') as f:
        for idx, param in enumerate(param_names):
            print(f'[{time.time()-t0:.2f}] {idx+1}/{len(param_names)}: process for {param_names[idx]}')
            f.write(f'Param: {param}\n')
            f.write(f'Effective Sample Size:{samples.getEffectiveSamples(idx)}\n')
            f.write(f'Gelman-Rubin (<<1) {samples.getGelmanRubin(idx)}\n')
            f.write('--------------------------------------------------\n')      
    
```
