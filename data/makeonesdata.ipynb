```python
import numpy as np

nsrcs = 8
nlens = 10
nbins = 20
ncombo = nsrcs*(nsrcs+1) + nlens*nsrcs + nlens
ndv = nbins*ncombo
print('ncombo = ', ncombo)
print('ndv = ', ndv)
```

```python
# produce ones mask
out = np.zeros((ndv,2))
out[:, 0] = np.arange(0, ndv)
out[:, 1] = 1
np.savetxt('/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/data/roman_ones.mask', out ,fmt=["%d", "%.6e"])
```

```python
# produce ones cov
out = np.zeros((int(ndv*(ndv+1)/2),10))
cnt = 0
for i in range(ndv):
    for j in range(i,ndv):
        out[cnt, 0] = i
        out[cnt, 1] = j
        if i==j:
            out[cnt, 8] = 1
        cnt+=1
print('cnt = ', cnt,'; len(out) = ', len(out)) 
        
np.savetxt('/project/chihway/junzhou/cocoa_approx/Cocoa/projects/roman_real/data/cov_roman_eyes', out ,fmt=["%d", "%d", "%.6e", "%.6e", "%d", "%d", "%d", "%d", "%.6e", "%.6e"])
```

### Do the same thing for LENS=SRCS

```python
import numpy as np

nsrcs = 8
nlens = 8
nbins = 20
ggl_exclude = [[6, 0], [7, 0], [7, 1]]
ncombo = nsrcs*(nsrcs+1) + nlens*nsrcs + nlens - len(ggl_exclude)
ndv = nbins*ncombo
print('ncombo = ', ncombo)
print('ndv = ', ndv)
```

```python
# produce ones mask
out = np.zeros((ndv,2))
out[:, 0] = np.arange(0, ndv)
out[:, 1] = 1
np.savetxt('./roman_les_ones.mask', out ,fmt=["%d", "%.6e"])
```

```python
# produce ones cov
out = np.zeros((int(ndv*(ndv+1)/2),10))
cnt = 0
for i in range(ndv):
    for j in range(i,ndv):
        out[cnt, 0] = i
        out[cnt, 1] = j
        if i==j:
            out[cnt, 8] = 1
        cnt+=1
print('cnt = ', cnt,'; len(out) = ', len(out)) 
        
np.savetxt('./cov_roman_les_eyes', out ,fmt=["%d", "%d", "%.6e", "%.6e", "%d", "%d", "%d", "%d", "%.6e", "%.6e"])
```
