```python
%config InlineBackend.figure_format = 'retina'
import sys, os
import numpy as np
import scipy as sp
sys.path.insert(0, os.path.realpath(os.path.join(os.getcwd(), '..')))
import getdist
from getdist import plots, loadMCSamples
import matplotlib.pyplot as plt
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
%matplotlib inline

def load(filepath, settings):
    samples = loadMCSamples(filepath, settings=settings)
    p = samples.getParams()
    h0 = p.H0/100
    try:
        samples.addDerived(
            h0,
            'h0',
            label = r'h_0',
        )
    except:
        True
    return samples

params = ['omegam', 'sigma8', 'ns', 'omegab', 'h0' ]
for i in range(8):
    params.append(f'roman_B1_{i+1}')
for i in range(8):
    params.append(f'roman_DZ_S{i+1}')
for i in range(8):
    params.append(f'roman_M{i+1}')
for i in range(2):
    params.append(f'roman_A1_{i+1}')
    
Dict = {
    'omegam':'\Omega_m',
    'sigma8':'\sigma_8',
    'ns'    :'n_s',
    'omegab':'\Omega_b',
    'h0'    :'h_0',
}
for i in range(8):
    Dict[f'roman_B1_{i+1}']  = f'b^{i+1}'
for i in range(8):
    Dict[f'roman_DZ_S{i+1}'] = f'\Delta_{{z}}^{i+1}'
for i in range(8):
    Dict[f'roman_M{i+1}']    = f'm_{i+1}'
Dict[f'roman_A1_1'] = f'A_{{\\rm IA}}'
Dict[f'roman_A1_2'] = f'\eta_{{\\rm IA}}'

with open('/project/chihway/junzhou/inference.results','w') as f:
    for domain in ['real', 'fourier']:
        
        samples = []
        for i in range(1,13):
            try:
                if i==1:
                    sample = load(f'/project/chihway/junzhou/cocoa/Cocoa/projects/roman_{domain}/chains/dc1_solution_maindatavector_MCMC', settings = {'ignore_rows':0.3, 'contours': [0.68]})
                    print(f'load {domain} main successfully')
                else:
                    sample = load(f'/project/chihway/junzhou/cocoa/Cocoa/projects/roman_{domain}/chains/dc1_solution_alternatesv{i}_MCMC', settings = {'ignore_rows':0.3, 'contours': [0.68]})
                    print(f'load {domain} v{i} successfully')
            except:
                if i==1:
                    print(f'fail to load {domain} main')
                    continue
                else:
                    print(f'fail to load {domain} v{i}')
                    continue
                
            samples.append(sample)
                
        for param in params:
            
            line = f'${Dict[param]}$ '
            f.write(line)
            
            for sample in samples:
                stats = sample.getMargeStats()
                p = stats.parWithName(param)
                line = f'& ${p.mean:.3f}^{{{p.limits[0].upper-p.mean:.3f}}}_{{{p.mean-p.limits[0].lower:.3f}}}$ '
                f.write(line)
                
            f.write(' \\\\ \n')
```
